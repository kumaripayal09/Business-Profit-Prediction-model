```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Sex=pd.get_dummies(customers['sex'])
Smoker=pd.get_dummies(customers['smoker'])
Region=pd.get_dummies(customers['region'])
```

```python
customers = pd.concat([customers,Sex,Smoker,Region],axis=1)
```

```python
customers.head(10)
```

```python
customers.drop(['sex','smoker','region'],axis=1)
```

```python
from sklearn import linear_model
```

```python
X=customers[['age','bmi','children','female','male','no','yes','northeast','northwest','southeast','southwest']]
```

```python
y=customers['charges']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.40, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
print('Accuracy of training subset: {:.3}'.format(lm.score(X_train,y_train)))
print('Accuracy of test subset:{:.3}'.format(lm.score(X_test,y_test)))
```
